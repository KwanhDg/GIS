# Phần 1: Tạo thư mục và import thư viện

```python
import os
os.makedirs('data/', exist_ok=True)  # Tạo thư mục data nếu chưa có
print("Thư mục data sẵn sàng!")

# Import thư viện cần thiết
import geopandas as gpd
import osmnx as ox
from shapely.geometry import box
import matplotlib.pyplot as plt

print("Import thành công!")
```

# Tải sông ngòi và kênh rạch từ OSM

```python
# Thiết lập OSMnx
ox.settings.use_cache = True  # Dùng cache để nhanh hơn lần sau
ox.settings.log_console = True  # Hiển thị tiến trình tải

# Bounding box HCMC: (west, south, east, north) – thứ tự của OSMnx
west, south, east, north = 106.45, 10.65, 106.95, 10.95
bbox = (west, south, east, north)

# Tags: waterway = river, canal, stream (sông chính + kênh rạch + suối nhỏ)
tags = {'waterway': ['river', 'canal', 'stream']}

# Tải features (đã là GeoDataFrame)
print("Đang tải dữ liệu từ OSM...")
rivers_gdf = ox.features.features_from_bbox(bbox, tags)

# Lọc an toàn: Chỉ giữ waterway hợp lệ (bỏ lọc 'highway' vì không tồn tại)
# Kiểm tra cột tồn tại trước khi dùng
if 'highway' in rivers_gdf.columns:
    rivers_gdf = rivers_gdf[rivers_gdf['highway'].isna()]  # Chỉ lọc nếu có cột này

# Giữ cột cần thiết (thêm kiểm tra cột)
keep_cols = ['geometry', 'name', 'waterway']
available_cols = [col for col in keep_cols if col in rivers_gdf.columns]
rivers_gdf = rivers_gdf[available_cols]

print(f"Số lượng sông/kênh tải về: {len(rivers_gdf)}")
print("\nPhân loại waterway:")
print(rivers_gdf['waterway'].value_counts())
print("\nTên một số sông chính:")
print(rivers_gdf['name'].dropna().unique()[:10])  # Ví dụ: Sông Sài Gòn, Kênh Tẻ...

# Lưu thành shapefile
rivers_gdf.to_file('data/hcmc_rivers.shp')
print("Đã lưu file: data/hcmc_rivers.shp")
```

# Preview bản đồ sông ngòi

```python
# Vẽ bản đồ đơn giản
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
rivers_gdf.plot(ax=ax, color='blue', linewidth=0.8, alpha=0.6)
ax.set_title('Sông ngòi và kênh rạch HCMC từ OSM')
ax.set_xlabel('Kinh độ (EPSG:4326)')
ax.set_ylabel('Vĩ độ (EPSG:4326)')
ax.set_xlim(west, east)
ax.set_ylim(south, north)
plt.tight_layout()
plt.show()

# Thống kê chiều dài chính xác: Re-project sang UTM 48N (mét)
projected_crs = 'EPSG:32648'  # UTM Zone 48N cho HCMC (miền Nam VN)
rivers_projected = rivers_gdf.to_crs(projected_crs)
total_length_m = rivers_projected.geometry.length.sum()  # Tổng chiều dài thực tế (mét)
total_length_km = total_length_m / 1000  # Chuyển sang km

print(f"Tổng chiều dài sông ước tính (chính xác): {total_length_km:.1f} km")
print(f"(So với ước tính cũ: {rivers_gdf.geometry.length.sum() * 111000 / 1000:.1f} km – sai số nhỏ)")
```

# Buffer quanh sông cho rủi ro ngập (từ Phần 3)

```python
from shapely.ops import unary_union
from shapely.geometry import box
import geopandas as gpd
import matplotlib.pyplot as plt

# Bước 1: Ghép tất cả sông thành một MultiLineString duy nhất
all_rivers = unary_union(rivers_gdf.geometry)

# Bước 2: Re-project sang CRS projected để buffer chính xác (mét thay vì độ)
projected_crs = 'EPSG:32648'  # UTM 48N cho HCMC
rivers_projected = rivers_gdf.to_crs(projected_crs)
all_rivers_projected = unary_union(rivers_projected.geometry)
rivers_buffer_projected = all_rivers_projected.buffer(100)  # Buffer 100m thực tế

# Bước 3: Re-project geometry Shapely đúng cách (sử dụng GeoSeries)
rivers_buffer_projected_gs = gpd.GeoSeries([rivers_buffer_projected], crs=projected_crs)
rivers_buffer = rivers_buffer_projected_gs.to_crs('EPSG:4326').iloc[0]  # Lấy geometry sau re-project

# Bước 4: Tạo GeoDataFrame đúng (chỉ geometry)
rivers_buffer_gdf = gpd.GeoDataFrame(geometry=[rivers_buffer], crs='EPSG:4326')
rivers_buffer_gdf['buffer_zone'] = 'river_buffer_100m'  # Cột tùy chọn

print(f"Buffer tạo thành công! Loại geometry: {rivers_buffer.geom_type}")
print(f"Diện tích buffer ước tính: {rivers_buffer.area * 111000**2 / 1e6:.1f} km² (xấp xỉ)")  # Ước tính từ độ
print(rivers_buffer_gdf.head())

# Lưu shapefile (bây giờ ổn)
rivers_buffer_gdf.to_file('data/hcmc_rivers_buffer.shp')
print("Đã lưu file: data/hcmc_rivers_buffer.shp")

# Preview buffer trên bản đồ (chồng với sông gốc)
west, south, east, north = 106.45, 10.65, 106.95, 10.95  # Bounding box
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
rivers_gdf.plot(ax=ax, color='blue', linewidth=0.8, alpha=0.6, label='Sông gốc')
rivers_buffer_gdf.plot(ax=ax, color='lightblue', alpha=0.4, edgecolor='darkblue', linewidth=1, label='Buffer 100m')
ax.set_title('Buffer quanh sông ngòi HCMC')
ax.legend()
ax.set_xlim(west, east)
ax.set_ylim(south, north)
plt.tight_layout()
plt.show()
```

# Xử lý file DEM

```python
import tarfile
import os
import rasterio
from rasterio.merge import merge
import matplotlib.pyplot as plt
from rasterio.plot import show

# Đường dẫn file .tar.gz
tar_path = 'data/rasters_COP30.tar.gz'  # Hoặc đường dẫn đầy đủ nếu ở Downloads

# Bước 1: Giải nén
extract_dir = 'data/dem_extracted/'
os.makedirs(extract_dir, exist_ok=True)
with tarfile.open(tar_path, 'r:gz') as tar:
    tar.extractall(path=extract_dir)
    print("Giải nén thành công!")
    print("Các file bên trong:")
    for member in tar.getmembers():
        if member.isfile() and member.name.endswith('.tif'):
            print(f"- {member.name} (DEM tile)")

# Bước 2: Tìm tất cả file .tif
tif_files = [os.path.join(extract_dir, f) for f in os.listdir(extract_dir) if f.endswith('.tif')]
print(f"\nSố tile DEM: {len(tif_files)}")

# Bước 3: Nếu chỉ 1 file, dùng trực tiếp; nếu nhiều, ghép mosaic
if len(tif_files) == 1:
    dem_path = tif_files[0]
    print(f"Dùng file đơn: {dem_path}")
else:
    print("Ghép mosaic các tile...")
    srcs = [rasterio.open(f) for f in tif_files]
    mosaic, out_trans = merge(srcs)
    profile = srcs[0].profile
    profile.update({"height": mosaic.shape[1], "width": mosaic.shape[2], "transform": out_trans})
    dem_path = 'data/hcmc_dem.tif'
    with rasterio.open(dem_path, 'w', **profile) as dst:
        dst.write(mosaic)
    for src in srcs:
        src.close()
    print(f"Đã ghép và lưu: {dem_path}")

# Bước 4: Kiểm tra DEM cuối cùng
with rasterio.open(dem_path) as src:
    dem = src.read(1)
    print(f"\nDEM cuối cùng:")
    print(f"- Shape: {dem.shape}")
    print(f"- CRS: {src.crs}")
    print(f"- Min/Max độ cao: {dem.min():.1f}m / {dem.max():.1f}m")

# Vẽ preview nhanh
plt.figure(figsize=(10, 8))
show(dem, cmap='terrain', title='Preview DEM HCMC')
plt.show()

print("DEM sẵn sàng! Giờ chạy Phần 2 để khám phá chi tiết.")
```

# Phần 2: Khám phá dữ liệu DEM (độ cao HCMC)

```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt

# Đường dẫn DEM (thay nếu tên file khác)
dem_path = 'data/dem_extracted/output_hh.tif'

# Đọc DEM
with rasterio.open(dem_path) as src:
    dem = src.read(1)  # Band 1 (độ cao)
    transform = src.transform  # Để vẽ extent
    crs = src.crs
    print(f"DEM info:")
    print(f"- Shape: {dem.shape} pixels (rows x cols)")
    print(f"- CRS: {crs}")
    print(f"- Độ phân giải: {src.res[0]:.1f}m x {src.res[1]:.1f}m")
    print(f"- Min/Max độ cao: {dem.min():.1f}m / {dem.max():.1f}m")

# Vẽ heatmap độ cao
plt.figure(figsize=(12, 10))
extent = [transform.c, transform.c + dem.shape[1] * transform.a,
          transform.f + dem.shape[0] * transform.e, transform.f]
im = plt.imshow(dem, cmap='terrain', extent=extent, interpolation='bilinear')
plt.colorbar(im, label='Độ cao (m)')
plt.title('Bản đồ độ cao DEM HCMC (Copernicus GLO-30)')
plt.xlabel('Kinh độ')
plt.ylabel('Vĩ độ')
plt.show()

# Thống kê vùng thấp dễ ngập (<5m)
low_elev_mask = dem < 5
low_pixels = np.sum(low_elev_mask)
pixel_area_m2 = src.res[0] * src.res[1]  # Diện tích mỗi pixel
low_area_km2 = (low_pixels * pixel_area_m2) / 1e6
total_area_km2 = (dem.shape[0] * dem.shape[1] * pixel_area_m2) / 1e6

print(f"\nThống kê rủi ro sơ bộ:")
print(f"- Diện tích vùng thấp (<5m): {low_area_km2:.2f} km² ({low_area_km2 / total_area_km2 * 100:.1f}% tổng diện tích)")
print(f"- Trung bình độ cao: {np.mean(dem):.1f}m")
print(f"- Độ lệch chuẩn: {np.std(dem):.1f}m")
```

# Tải ranh giới quận HCMC từ GADM level 3

```python
import geopandas as gpd
import os

# Đường dẫn thư mục GADM
gadm_dir = 'data/gadm_vn/'
shp_file = os.path.join(gadm_dir, 'gadm41_VNM_3.shp')  # Level 3 (quận/huyện)

# Load toàn bộ VN level 3
if os.path.exists(shp_file):
    admin_gdf = gpd.read_file(shp_file)
    print(f"Tổng đơn vị hành chính VN level 3: {len(admin_gdf)}")

    # Kiểm tra tên tỉnh HCMC (có thể 'Hồ Chí Minh City' hoặc 'Ho Chi Minh')
    hcmc_names = admin_gdf['NAME_1'].unique()
    print("Các tên tỉnh có 'HCMC':", [name for name in hcmc_names if 'Hồ Chí Minh' in name or 'Ho Chi Minh' in name])

    # Lọc HCMC (thay 'Hồ Chí Minh City' nếu tên khác)
    hcmc_gdf = admin_gdf[admin_gdf['NAME_1'] == 'Hồ Chí Minh City'].to_crs('EPSG:4326')
    if len(hcmc_gdf) == 0:
        hcmc_gdf = admin_gdf[admin_gdf['NAME_1'] == 'Ho Chi Minh'].to_crs('EPSG:4326')  # Thử tên thay thế

    print(f"Số quận/huyện HCMC: {len(hcmc_gdf)}")
    print("Tên vài quận:", hcmc_gdf['NAME_3'].head().tolist()[:5])  # Ví dụ: Quận 1, Quận 3...

    # Lưu chỉ HCMC để dùng sau
    hcmc_gdf.to_file('data/hcmc_districts.shp')
    print("Đã lưu: data/hcmc_districts.shp")
else:
    print("File gadm41_VNM_3.shp không tìm thấy trong data/gadm_vn/ – kiểm tra tên file.")
```

# Phần 3: Phân tích rủi ro ngập (DEM + buffer sông + quận HCMC)

```python
import rasterio
from rasterio.mask import mask
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt

# Load dữ liệu (cập nhật dem_path)
dem_path = 'data/dem_extracted/output_hh.tif'  # DEM mới
rivers_buffer_gdf = gpd.read_file('data/hcmc_rivers_buffer.shp')
# hcmc_gdf từ cell tải GADM trước

# Fix geometry invalid trong hcmc_gdf (buffer(0) để sửa tự động)
hcmc_gdf['geometry'] = hcmc_gdf['geometry'].buffer(0)
hcmc_gdf = hcmc_gdf[~hcmc_gdf.is_empty]  # Chỉ bỏ empty, không valid strict
print(f"Số quận sau fix và lọc: {len(hcmc_gdf)}")

if len(hcmc_gdf) == 0:
    print("Lỗi: hcmc_gdf rỗng! Kiểm tra cell tải GADM, đảm bảo tên 'NAME_1' đúng.")
    # Vẽ bản đồ ranh giới thô để debug
    hcmc_gdf_raw = gpd.read_file('data/hcmc_districts.shp')  # Nếu có file lưu trước
    hcmc_gdf_raw.plot(figsize=(10, 8))
    plt.title('Debug: Ranh giới HCMC thô')
    plt.show()
else:
    # Bước 1: Mô phỏng ngập từ DEM (vùng < flood_level, shape đầy đủ)
    flood_level = 2.0  # m
    with rasterio.open(dem_path) as src:
        dem = src.read(1)
        flood_mask = (dem <= flood_level).astype(np.uint8)  # 1=ngập, 0=không

    print(f"Shape flood_mask: {flood_mask.shape}")

    # Bước 2: Mask vùng buffer sông trên DEM đầy đủ (crop=False để giữ shape)
    with rasterio.open(dem_path) as src:
        out_image, out_transform = mask(src, [rivers_buffer_gdf.geometry.iloc[0]], crop=False, nodata=-9999)
        river_area = (out_image[0] != -9999).astype(np.uint8)  # 1=trong buffer sông, 0=ngoài

    # Kết hợp: Ngập DEM HOẶC trong buffer sông (shape khớp)
    combined_mask = np.maximum(flood_mask, river_area)  # OR logic: 1 nếu ngập hoặc gần sông
    print(f"Shape combined_mask: {combined_mask.shape}")

    # Bước 3: Loop từng quận để tính stats (mean elev, % ngập)
    mean_elev_list = []
    min_elev_list = []
    flood_pct_list = []
    for idx, row in hcmc_gdf.iterrows():
        district_geom = [row.geometry]
        try:
            with rasterio.open(dem_path) as src:
                out_dem, _ = mask(src, district_geom, crop=False, nodata=-9999)
                district_dem = out_dem[0]
                valid_mask = district_dem != -9999
                if np.sum(valid_mask) > 0:
                    mean_elev = np.mean(district_dem[valid_mask])
                    min_elev = np.min(district_dem[valid_mask])
                else:
                    mean_elev = np.nan
                    min_elev = np.nan

                # % ngập trong quận
                district_flood = combined_mask[valid_mask]
                flood_pct = (np.sum(district_flood) / len(district_flood)) * 100 if len(district_flood) > 0 else 0
        except Exception as e:
            print(f"Lỗi mask quận {row['NAME_3']}: {e}")
            mean_elev = np.nan
            min_elev = np.nan
            flood_pct = 0

        mean_elev_list.append(mean_elev)
        min_elev_list.append(min_elev)
        flood_pct_list.append(flood_pct)

    # Gán vào hcmc_gdf
    hcmc_gdf['mean_elev'] = mean_elev_list
    hcmc_gdf['min_elev'] = min_elev_list
    hcmc_gdf['flood_pct'] = flood_pct_list
    hcmc_gdf['risk_level'] = np.where(hcmc_gdf['flood_pct'] > 20, 'Cao', 
                                      np.where(hcmc_gdf['flood_pct'] > 10, 'Trung bình', 'Thấp'))

    # Hiển thị bảng (top 10 quận ngập cao nhất)
    print("Bảng rủi ro ngập theo quận HCMC (top 10):")
    result_df = hcmc_gdf[['NAME_3', 'mean_elev', 'flood_pct', 'risk_level']].sort_values('flood_pct', ascending=False).round(2).head(10)
    print(result_df)

    # Vẽ bản đồ % ngập (aspect='equal')
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    hcmc_gdf.plot(column='flood_pct', cmap='Reds', legend=True, ax=ax, edgecolor='black', linewidth=0.5, aspect='equal')
    plt.title('% Diện tích rủi ro ngập theo quận HCMC')
    plt.axis('off')
    plt.show()
```
